```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
gdp_file = "GDP By Country.csv"
gdp_df = pd.read_csv(gdp_file)
gdp_df.head()
```

```python
gdp_new = gdp_df.drop(columns=['Country Code','Indicator Name','Indicator Code'])
gdp_new.drop('2018',axis=1, inplace=True)
```

```python
rename = gdp_new.rename(columns={'Country Name':'Country'})
rename2 = rename.set_index('Country')
rename3 = rename2.add_suffix(' gdp')
```

```python
clean_gdp = rename3.dropna()
final_gdp = clean_gdp.drop(clean_gdp.columns.to_series()["1960":"1989"],axis=1)

```

```python
final_gdp.to_csv('clean_gdp_file.csv')
```
